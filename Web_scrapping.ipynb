# show The line charts of total runs in each over 

```python
sun=matches[matches['batting_team']=='Sunrisers Hyderabad']
rcb=matches[matches['batting_team']=='Royal Challengers Bangalore']
```

```python
sun.head()
```

```python
sun.shape
```

```python
rcb.shape
```

```python
total_runs=pd.DataFrame(rcb.groupby('over')['total_runs'].sum()).reset_index() 
total_runs.rename(columns={'total_runs':'rcb'},inplace=True)
```

```python
total_runs
```

```python
total_runs['sun']=sun.groupby('over')['total_runs'].sum().values
```

```python
total_runs
```

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['rcb'],marker='<',color='green',markersize=8)
plt.xticks(total_runs['over'])

for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['rcb'].tolist()[i],total_runs['rcb'].tolist()[i],fontsize=12, fontweight='bold')


plt.title('over vs total runs')
plt.xlabel('overs')
plt.ylabel('runs')

plt.show()
```

```python
total_runs['over'].tolist()
```

```python
total_runs['over'].tolist()[0]
```

```python
total_runs['rcb'].tolist()
```

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['sun'],marker='<',color='green',markersize=8)
plt.plot(total_runs['over'],total_runs['rcb'],marker='<',color='red',markersize=8)
plt.xticks(total_runs['over'])

for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['sun'].tolist()[i],total_runs['sun'].tolist()[i],fontsize=12, fontweight='bold')
    plt.text(total_runs['over'].tolist()[i],total_runs['rcb'].tolist()[i],total_runs['rcb'].tolist()[i],fontsize=12, fontweight='bold')


plt.title('over vs total runs')
plt.xlabel('overs')
plt.ylabel('runs')

plt.show()
```

# Comparing both teams overs and runs 

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['sun'],marker='<',color='green',markersize=8)

plt.xticks(total_runs['over'])

for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['sun'].tolist()[i],total_runs['sun'].tolist()[i],fontsize=12, fontweight='bold')


plt.title('over vs total runs')
plt.xlabel('overs')
plt.ylabel('runs')

plt.show()
```

```python
total_runs['rcb_rolling']=total_runs['rcb'].rolling(window=3).mean()
total_runs['sun_rolling']=total_runs['sun'].rolling(window=3).mean()
```

```python
total_runs
```

# Rolling means 

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['rcb_rolling'],marker='<',color='green',markersize=8)
plt.xticks(total_runs['over'])

for i in range(len(total_runs['over'].tolist())):
    plt.text(total_runs['over'].tolist()[i],total_runs['rcb_rolling'].tolist()[i],total_runs['rcb'].tolist()[i],fontsize=12, fontweight='bold')


plt.title('over vs total runs')
plt.xlabel('overs')
plt.ylabel('runs')

plt.show()
```

# Cumulative Sums 

```python
total_runs
```

```python
total_runs['rcb_cum']=total_runs['rcb'].cumsum()
total_runs['sun_cum']=total_runs['sun'].cumsum()
```

```python
total_runs
```

```python
plt.figure(figsize=(12,4))
plt.plot(total_runs['over'],total_runs['rcb_cum'],marker='<',color='green',markersize=8,label='rcb')
plt.plot(total_runs['over'],total_runs['sun_cum'],marker='+',color='red',markersize=8,label='sun')
plt.xticks(total_runs['over'])

plt.title('over vs total runs')
plt.xlabel('overs')
plt.ylabel('runs')
plt.grid()
plt.legend()

plt.show()
```
